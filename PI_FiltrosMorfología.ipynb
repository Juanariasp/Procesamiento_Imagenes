```python
import cv2
import PIL
import numpy as np
import plotly.subplots as sp
import plotly.graph_objects as go

from PIL import Image
import scipy.signal as sig
from scipy.ndimage import generic_filter
```

# **Problema 2**

```python
url = "/content/drive/MyDrive/Colab Notebooks/Procesamiento de Imagenes/Imagenes/cameraman-noise.tif"
```

```python
img = Image.open(url)
img = img.convert("L")
```

```python
#Se definen las dimensiones para la imagen.
img = np.asarray(img)
wid, hgt = img.shape[1], img.shape[0]
print("Ancho de la imagen: ", wid)
print("Alto de la imagen: ", hgt)
```

```python
fig = go.Figure(data=go.Heatmap(z=img, colorscale='gray'))

fig.update_layout(yaxis=dict(autorange="reversed"))

fig.update_layout(
    title={
        'text':'Imagen Original',
        'x':0.5}, 
    height=wid*2, width=wid*2,
    margin=dict(l=10, r=10, t=35, b=10)
)

fig.show()
```

# **Realize una convoluci√≥n con un box filter (lleno de unos) para reducir el ruido**

```python
mascara_unos = np.array([[1, 1, 1], [1, 1, 1], [1, 1, 1]])
img_m3 = sig.convolve2d(img, mascara_unos, mode='same')
```

```python
fig = go.Figure(data=go.Heatmap(z=img_m3, colorscale='gray'))

fig.update_layout(yaxis=dict(autorange="reversed"))

fig.update_layout(
    title={
        'text':'Imagen con mascara de unos',
        'x':0.5}, 
    height=wid*2, width=wid*2,
    margin=dict(l=10, r=10, t=35, b=10)
)

fig.show()
```

```python
fig = sp.make_subplots(rows=1, cols=2)

#Imagen Original
fig.add_trace(
    go.Heatmap(z=img, colorscale='gray', showscale=True), row=1, col=1
)

#Imagen con mascara 3x3
fig.add_trace(
    go.Heatmap(z=img_m3, colorscale='gray', showscale=False), row=1, col=2,
)

fig.update_layout(
    title={
        'text':'Resultado de los filtros',
        'x':0.5}, 
    height=512, width=1024,
    margin=dict(l=10, r=10, t=35, b=10),
    yaxis=dict(autorange="reversed"), 
    yaxis2=dict(autorange="reversed"),
)

fig.show()
```

# **Seleccione un tama√±o razonable del filtro para compensar el ruido y el suavizado**

```python
mascara_p = np.ones((5,5), dtype=np.float32)
img_m5 = sig.convolve2d(img, mascara_p, mode='same')
```

```python
fig = go.Figure(data=go.Heatmap(z=img_m5, colorscale='gray'))

fig.update_layout(yaxis=dict(autorange="reversed"))

fig.update_layout(
    title={
        'text':'Imagen con mascara de unos',
        'x':0.5}, 
    height=wid*2, width=wid*2,
    margin=dict(l=10, r=10, t=35, b=10)
)

fig.show()
```

```python
fig = sp.make_subplots(rows=1, cols=2)

#Imagen Original
fig.add_trace(
    go.Heatmap(z=img, colorscale='gray', showscale=True), row=1, col=1
)

#Imagen con mejor mascara 
fig.add_trace(
    go.Heatmap(z=img_m5, colorscale='gray', showscale=False), row=1, col=2,
)

fig.update_layout(
    title={
        'text':'Resultado de los filtros',
        'x':0.5}, 
    height=512, width=1024,
    margin=dict(l=10, r=10, t=35, b=10),
    yaxis=dict(autorange="reversed"), 
    yaxis2=dict(autorange="reversed"),
)

fig.show()
```

# **Aplique un filtro mediana con una m√°scara de 3x3 y compare los resultados con el punto 3.**

```python
img_mediana = sig.medfilt2d(img, (3,3))
```

```python
fig = go.Figure(data=go.Heatmap(z=img_mediana, colorscale='gray'))

fig.update_layout(yaxis=dict(autorange="reversed"))

fig.update_layout(
    title={
        'text':'Imagen con mascara de unos',
        'x':0.5}, 
    height=wid*2, width=wid*2,
    margin=dict(l=10, r=10, t=35, b=10)
)

fig.show()
```

```python
fig = sp.make_subplots(rows=1, cols=3)

#Imagen Original
fig.add_trace(
    go.Heatmap(z=img, colorscale='gray', showscale=True), row=1, col=1
)

#Imagen con mascara 5x5
fig.add_trace(
    go.Heatmap(z=img_m5, colorscale='gray', showscale=False), row=1, col=2,
)

#Imagen con filtro mediana
fig.add_trace(
    go.Heatmap(z=img_mediana, colorscale='gray', showscale=False), row=1, col=3,
)

fig.update_layout(
    title={
        'text':'Resultado de los filtros',
        'x':0.5}, 
    height=400, width=1100,
    margin=dict(l=10, r=10, t=35, b=10),
    yaxis=dict(autorange="reversed"), 
    yaxis2=dict(autorange="reversed"),
    yaxis3=dict(autorange="reversed")
)

fig.show()
```

# **¬øCu√°l aproximaci√≥n es mas satisfactoria en terminos de calidad de la imagen?**


<p align="Justify">
Se puede concluir que la imagen filtrada con el filtro de mediana tiene una mejor calidad en t√©rminos de reducci√≥n de ruido en comparaci√≥n con la imagen filtrada solo con una m√°scara de unos.<br><br>
Esto se debe a que el filtro de mediana es m√°s efectivo en eliminar el ruido de la imagen, ya que utiliza la mediana de los valores de los p√≠xeles en lugar de su promedio, lo que permite preservar mejor las caracter√≠sticas importantes de la imagen y evitar la p√©rdida de detalles.<br><br>
Por lo tanto, al utilizar el filtro de mediana se logra reducir el ruido de la imagen sin afectar significativamente su calidad, lo que resulta en una imagen m√°s n√≠tida y clara en comparaci√≥n con el filtro de mascara lleno de unos que puede producir una imagen borrosa y con p√©rdida de detalles.
